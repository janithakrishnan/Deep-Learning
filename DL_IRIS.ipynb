KERAS SEQUENTIAL MODEL FOR IRIS DATASET CLASSIFICATION

LOADING IRIS DATASET FROM SCIKIT LEARN

```python
from sklearn.datasets import load_iris
iris=load_iris()
X=iris.data
y=iris.target
print(X.shape)
print(y.shape)
```

```python
print(y)
```

SPLITTING THE DATA TO TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
```

CREATING,TRAINING AND TESTING DEEP LEARNING MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
dl_model=Sequential() #Initialising the model
dl_model.add(Dense(4,input_dim=4,activation='relu')) #adding first hidden layer with 4 neurons and reLu as activation function
dl_model.add(Dense(3,activation='softmax')) #adding 3 neuron output layer (since there are 3 classes) and softmax as activation fn since IRIS is a multiclass problem
dl_model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy']) #configures the training model
iris_history=dl_model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,verbose=0) #fits and validate the model and stores the resultant metrics in the object "iris_history"
```

PRINTING THE EPOCH HISTORY

```python
print("Training Accuracy:",iris_history.history['accuracy'])
print("Training Loss:",iris_history.history['loss'])
print("Testing Accuracy:",iris_history.history['val_accuracy'])
print("Testing Loss:",iris_history.history['val_loss'])
```

PLOTTING THE ACCURACY

```python
import matplotlib.pyplot as plt
plt.plot(iris_history.history['accuracy'],label='Training Accuracy')
plt.plot(iris_history.history['val_accuracy'],label='Testing Accuracy')
plt.legend()
plt.show()
```

PLOTTING LOSS

```python
plt.plot(iris_history.history['loss'],label='Training Loss')
plt.plot(iris_history.history['val_loss'],label='Testing Loss')
plt.legend()
plt.show()
```
