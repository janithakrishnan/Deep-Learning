KERAS SEQUENTIAL MODEL FOR BREAST CANCER DATASET CLASSIFICATION - BINARY CLASSIFICATION PROBLEM

*LOADING* BREAST CANCER DATASET FROM SCIKIT LEARN

```python
from sklearn.datasets import load_breast_cancer
bc=load_breast_cancer()
X=bc.data
y=bc.target
print(X.shape)
print(y.shape)
```

```python
print(y)
```

SPLITTING THE DATA TO TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
```

CREATING,TRAINING AND TESTING DEEP LEARNING MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
dl_model=Sequential() #Initialising the model
dl_model.add(Dense(8,input_dim=30,activation='relu')) #adding first hidden layer with 4 neurons and reLu as activation function
dl_model.add(Dense(1,activation='sigmoid')) #1 neuron output layer and sigmoid as activation fn since BREAST CANCER is a binary classification problem
dl_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) #configures the training model
bc_history=dl_model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,verbose=0) #fits and validate the model and stores the resultant metrics in the object "iris_history"
```

PRINTING THE EPOCH HISTORY

```python
print("Training Accuracy:",bc_history.history['accuracy'])
print("Training Loss:",bc_history.history['loss'])
print("Testing Accuracy:",bc_history.history['val_accuracy'])
print("Testing Loss:",bc_history.history['val_loss'])
```

PRINTING MAXIMUM TRAINING AND TESTING ACCURACIES

```python
print("Maximum Training Accuracy",bc_history.history['accuracy'][-1])
print("Maximum Testing Accuracy",bc_history.history['val_accuracy'][-1])
```

PLOTTING THE ACCURACY

```python
import matplotlib.pyplot as plt
plt.plot(bc_history.history['accuracy'],label='Training Accuracy')
plt.plot(bc_history.history['val_accuracy'],label='Testing Accuracy')
plt.title('Breast Cancer Classification Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

PLOTTING LOSS

```python
plt.plot(bc_history.history['loss'],label='Training Loss')
plt.plot(bc_history.history['val_loss'],label='Testing Loss')
plt.title('Breast Cancer Classification Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
